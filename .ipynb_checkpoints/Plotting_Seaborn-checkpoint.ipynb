Plotting examples with Seaborn

gallary: https://seaborn.pydata.org/examples/index.html

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
#%matplotlib inline
```

```python
#load dummy dataset
df=pd.read_csv('Pokemon.csv', index_col=0)
df.head()
```

Scatterplot

```python
sns.lmplot(x='Attack', y='Defense', data=df)
```

This is doing regression, if we just want the scatterplot, remove this line, color by stage of pokemon

```python
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue='Stage')
```

Now, use matplotlib's customization functions to set axis limits


```python
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue='Stage')
plt.ylim(0,200)
```

## Histogram

only want a subset of the data

```python
stats_df=df.drop(['Total', 'Stage', 'Legendary'], axis=1)

sns.boxplot(data=stats_df)
```

## Violoen plot
shows distribution by thickness, the thicker it is the more datapoints are at those values.

also change the default style to whitegrid

```python
#set theme
sns.set_style('whitegrid')


#set custom colors in the order of plotting
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]


#set plot size ahead of time with matplot lib
fig, ax = plt.subplots(figsize=(12,8))
sns.violinplot(x='Type 1', y='Attack', data=df, palette=pkmn_type_colors)

```

## Overlaying Plots

1. plot violen plot
2. plot swarmplot over it

```python
#set figure size

plt.figure(figsize=(12,8))

# Create plot
sns.violinplot(x='Type 1',
               y='Attack', 
               data=df, 
               inner=None, # Remove the bars inside the violins
               palette=pkmn_type_colors)
 
sns.swarmplot(x='Type 1', 
              y='Attack', 
              data=df, 
              color='k', # Make points black
              alpha=0.7) # and slightly transparent
```

## Melt the data from wide to long format

ID variable = variable to keep, all others will be melted. we want to keep Name, type1, and type2

```python
melted_df=pd.melt(stats_df, id_vars=['Name', 'Type 1', 'Type 2'], var_name='Stat')
melted_df.head()
```

```python
sns.boxplot(data=melted_df, x='Type 1', y='value', hue='Stat')
```

```python
## factorplot
```

```python
plt.figure(figsize=(25,20))
sns.set(font_scale=3)
g=sns.factorplot(data=df, x="Type 1", y= "Attack", col = "Stage", margin_titles=True, height=15, kind='box')

# Rotate x-axis labels
g.set_xticklabels(rotation=-45)
```

## Barplot

```python
# Count Plot (a.k.a. Bar Plot)
sns.countplot(x='Type 1', data=df, palette=pkmn_type_colors) #counts up the frequency in each category
 
# Rotate x-labels
plt.xticks(rotation=-45)
```

## Joint Distributions
Joint distribution plots combine information from scatter plots and histograms to give you detailed information for bi-variate distributions.

```python
sns.set(font_scale=1.1)
sns.jointplot(x='Attack', y='Defense', data=df)
```
