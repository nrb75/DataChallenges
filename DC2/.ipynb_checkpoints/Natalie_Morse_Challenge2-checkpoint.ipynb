## Data Challenge #2

## Goal
Pricing optimization is, non surprisingly, another area where data science can provide huge value.

The goal here is to evaluate whether a pricing test running on the site has been successful. As always, you should focus on user segmentation and provide insights about segments who behave differently as well as any other insights you might find.

## Challenge Description
Company XYZ sells a software for $39. Since revenue has been flat for some time, the VP of Product has decided to run a test increasing the price. She hopes that this would increase revenue. In the experiment, 66% of the users have seen the old price ($39), while a random sample of 33% users a higher price ($59).

The test has been running for some time and the VP of Product is interested in understanding how it went and whether it would make sense to increase the price for all the users. Especially she asked you the following questions:

1. Should the company sell its software for $39 or $59?

2. The VP of Product is interested in having a holistic view into user behavior, especially focusing on actionable insights that might increase conversion rate. What are your main findings looking at the data?

3. The VP of Product feels that the test has been running for too long and she should have been able to get statistically significant results in a shorter time. Do you agree with her intuition? After how many days would you have stopped the test? Please, explain why.


## Data
We have two tables downloadable by clicking here. The two tables are:

### "test_results" - data about the test
Columns:

user_id : the Id of the user. Can be joined to user_id in user_table

timestamp : the date and time when the user hit for the first time company XYZ webpage. It is in user local time

source : marketing channel that led to the user coming to the site. It can be: ads-["google", "facebook", "bing", "yahoo", "other"]. That is, user coming from google ads, yahoo ads, etc. seo - ["google", "facebook", "bing", "yahoo", "other"]. That is, user coming from google search, yahoo, facebook, etc.

friend_referral : user coming from a referral link of another user

direct_traffic: user coming by directly typing the address of the site on the browser

device : user device. Can be mobile or web

operative_system : user operative system. Can be: "windows", "linux", "mac" for web, and "android", "iOS" for mobile. "Other" if it is none of the above

test: whether the user was in the test (i.e. 1 -> higher price) or in control (0 -> old, lower price)

price : the price the user sees. It should match test

converted : whether the user converted (i.e. 1 -> bought the software) or not (0 -> left the site without buying it).

### "user_table" - Information about the user
Columns:
user_id : the Id of the user. Can be joined to user_id in test_results table

city : the city where the user is located. Comes from the user ip address

country : in which country the city is located

lat : city latitude - should match user city

long : city longitude - should match user city

## 1. Should the company sell its software for $39 or  $59?

I recommend that they sell their product for $59, even though the conversion rate is lower at this value, you make up that revenue by having a higher priced product.

Approach: compare the test and control groups conversion rates, start with overall figures

```python
buy_rate_all=df_test_clean.groupby('test', as_index=False)['converted'].count()
buy_rate_all.rename(columns={'converted': 'total'}, inplace=True)
buy_rate_converted=df_test_clean.groupby('test', as_index=False)['converted'].sum()
buy_rate_all=pd.merge(buy_rate_all, buy_rate_converted, on='test')
buy_rate_all['percent']=buy_rate_all['converted']/buy_rate_all['total']
```

```python
buy_rate_all
```

```python
fig, ax1  = plt.subplots(1, 1, figsize=(7, 5), sharex=True)
#fig.canvas.draw()
sns.barplot(x=buy_rate_all['test'], y=buy_rate_all['percent'], palette="rocket", ax=ax1)
ax1.axhline(0, color="k", clip_on=False)
ax1.set_ylabel("Conversion rate (%)")
ax1.set_xlabel("Test or Control Group")
#plt.xticks([0,1], ('Random Model', 'FP Model'))
```

#### Because these groups are unbalanced, more people were in the control (0) group compared to the test (1) group, let's see if exactly the same amount of people were in each group, and then apply these conversion rates to see differences in sales

```python
#hypothetical sales if 1000 people in each group
control_price=39
test_price=59

sales_control=buy_rate_all['percent'][0]*control_price*1000 #for 1000 ppl
sales_test=buy_rate_all['percent'][1]*test_price*1000
```

```python
#sales for control group with 1000 people
sales_control
```

```python
#sales for test group with 1000 people
sales_test
```

Even though the conversion rate is slightly lower for the test group, because the product is selling at a higher price we ar still able to increase sales over the control group, if the same number of people are in each group.

### Statistical tests between the 2 groups:
Since this is a binary trial (purchased = success(1), no purhcase = fail (0), I will test the differences between the groups with a z-test.

```python
buy_rate_all
```

```python
import statsmodels
from statsmodels.stats.proportion import proportions_ztest

proportions_ztest(count=buy_rate_all.converted, nobs=buy_rate_all.total, alternative='two-sided')
```

The first value is the z-statistic, and the 2nd value is the p-value. Because the p-value is so small, (<0.001), I conclude that the conversion rates for these 2 groups are not the same.

The conversion rate is higher in the control group.

## 3. The VP of Product feels that the test has been running for too long and she should have been able to get statistically significant results in a shorter time. Do you agree with her intuition? After how many days would you have stopped the test? Please, explain why.


## Plot the conversion rate over time

```python
buy_rate_day=df_test_clean.groupby('date', as_index=False)['converted'].count()
buy_rate_day.rename(columns={'converted' :'total'}, inplace=True)

buy_rate_converted=df_test_clean.groupby('date', as_index=False)['converted'].sum()#counts the 1s up as total ppl who bought product

buy_rate_converted=pd.merge(buy_rate_converted, buy_rate_day, on='date')
buy_rate_converted['percent']=buy_rate_converted['converted']/buy_rate_converted['total']
```

```python
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter

fig, ax = plt.subplots()
ax.plot(buy_rate_converted['date'], buy_rate_converted['percent'])

myFmt = DateFormatter("%m-%d")
ax.xaxis_date()
ax.xaxis.set_major_formatter(myFmt)

## Rotate date labels automatically
fig.autofmt_xdate()
plt.ylabel('Conversion rate (%)')
plt.xlabel('month-day')
plt.title('Percent of customers who purchased product')
plt.show()
```

```python
#conversion rate over time split by test/control groups
buy_rate_group=df_test_clean.groupby(['date', 'test'], as_index=False)['converted'].count()
buy_rate_group.rename(columns={'converted' :'total'}, inplace=True)

buy_rate_group_con=df_test_clean.groupby(['date', 'test'], as_index=False)['converted'].sum()#counts the 1s up as total ppl who bought product
buy_rate_group_con=pd.merge(buy_rate_group, buy_rate_group_con, on=['date', 'test'])
buy_rate_group_con['percent']=buy_rate_group_con['converted']/buy_rate_group_con['total']
```

```python
buy_rate_group_con['test']=np.where(buy_rate_group_con['test']==1, 'test', 'control')

#buy_rate_group_con['test'].where(buy_rate_group_con['test_x']=1
buy_rate_group_con.head()
```

```python
grouped=buy_rate_group_con.groupby('test')
colors = {'test':'red', 'control':'blue'}

fig, ax = plt.subplots()

for key, group in grouped:
    group.plot(ax=ax, kind='line', x='date', y='percent', label=key, color=colors[key])

plt.show()
```

The conversion rate, or the % of customers who buy the product is consistently higher in the Control group than the Test group over time.


It appears there is some variability in the data over time, with some peaks in conversion rates happening throughout the data. However, this does not seem to be time dependent and instead looks like a fairly random phenomenon. Therefore, it does seem likely that the test could be stopped sooner and still produce the same results. I suggest the test be stopped after 1 month, because this is enough time to capture the variability in the data.

I will test this 1 month window using the same methods as earlier (z-test) between the test and control groups.

```python
min(df_test_clean['date'])+30
```

```python
df_test_clean30=df_test_clean[df_test_clean['date']<'2015-04-03']
buy_rate_all=df_test_clean30.groupby('test', as_index=False)['converted'].count()
buy_rate_all.rename(columns={'converted': 'total'}, inplace=True)
buy_rate_converted=df_test_clean30.groupby('test', as_index=False)['converted'].sum()
buy_rate_all=pd.merge(buy_rate_all, buy_rate_converted, on='test')
buy_rate_all['percent']=buy_rate_all['converted']/buy_rate_all['total']

fig, ax1  = plt.subplots(1, 1, figsize=(7, 5), sharex=True)
#fig.canvas.draw()
sns.barplot(x=buy_rate_all['test'], y=buy_rate_all['percent'], palette="rocket", ax=ax1)
ax1.axhline(0, color="k", clip_on=False)
ax1.set_ylabel("Conversion rate (%)")
ax1.set_xlabel("Test or Control Group")
#plt.xticks([0,1], ('Random Model', 'FP Model'))
```

This plot is very simmilar to the plot with all the data, and statistical tests are also likely to show the entire dataset and 30-day datset to be fairly equal.
